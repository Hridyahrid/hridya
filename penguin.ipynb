```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your DataFrame (assuming it's already preprocessed)
penguins_size = pd.read_csv('penguins_size.csv')

plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.plot(penguins_size['culmen_length_mm'], penguins_size['flipper_length_mm'], marker='o', linestyle='-')
plt.title('Flipper Length vs Culmen Length') # Changed title to reflect the new plot
plt.xlabel('Culmen Length (mm)') # Changed x-axis label
plt.ylabel('Flipper Length (mm)')
plt.grid(True)
plt.show()

# Example 2: Line plot of body mass by species (remains unchanged)
species_groups = penguins_size.groupby('species')
plt.figure(figsize=(10, 6))
for species, group in species_groups:
    plt.plot(group['culmen_length_mm'], group['body_mass_g'], marker='o', linestyle='-', label=species)
plt.title('Body Mass by Species and Culmen Length')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Body Mass (g)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your DataFrame (assuming it's already preprocessed)
penguins_size = pd.read_csv('penguins_size.csv')
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.scatter(penguins_size['culmen_length_mm'], penguins_size['flipper_length_mm'],
            c=penguins_size['body_mass_g'], cmap='viridis', alpha=0.7)  # Color by body mass
plt.title('Flipper Length vs. Culmen Length')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Flipper Length (mm)')
plt.colorbar(label='Body Mass (g)')  # Add colorbar for body mass
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
penguins_size = pd.read_csv('penguins_size.csv')
plt.figure(figsize=(8, 6))
sns.histplot(data=penguins_size, x='flipper_length_mm', bins=20, hue='species', palette='viridis')
plt.title('Distribution of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
